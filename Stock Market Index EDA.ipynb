Now convert the whole Dataframe to a Year Period Index

Take note that each respective column will use different functions (min, max, sum, last, first)

For example, the Highest bid price should use *max* for that Year Period
While the Trade Volume, should be the *sum* for all the dates (days + months aggregated together)

**Columns needed**
- Open
- High
- Low
- Close
- Adj Close
- Volume

```python
# First Create an empty Dataframe

year_sp = pd.DataFrame()
year_sp
```

```python
## Next append to the Dataframe the respective columns
## Create Open Column and append with the relevant data from the dataframe = sp

year_sp['Open'] = sp['Open'].resample('Y').first()
year_sp
```

```python
## Finish with the rest
year_sp['High'] = sp.High.resample('Y').max()
year_sp['Low'] = sp.Low.resample('Y').min()
year_sp['Close'] = sp.Close.resample('Y').last()
year_sp['Volume'] = sp.Volume.resample('Y').sum()
year_sp.head()
```

Create a new Axes with an invisible x-axis and an independent y-axis positioned opposite to the original one (i.e. at right). The x-axis autoscale setting will be inherited from the original Axes. To ensure that the tick marks of both y-axes align, see LinearLocator.

## Adding Events to Plots
### Marking out respective x labels

```python
# Define events
# A list of tuples
events = [
    (datetime(2017,8,11), 'Cream Cream'),
    (datetime(2019,3,11), 'Banila'),
    (datetime(2019,11,11), 'KWT'),

]
# Axes
fig = plt.figure( figsize=(16,6), dpi=100 )
ax = fig.add_subplot(1,1,1)

# Data Values
x = new_spx.index
y1 = new_spx.values
y2 = new_nik.values

## ax.plot and plt.plot is the same
ax.plot(x,y1, color=dt_hex, linewidth = 3, linestyle = 'solid', label = 'S&P 500')

#Use np.vstack to stack two ndarray vertically
both_y = np.vstack([y1,y2])
ax.set_ylim(both_y.min()*0.90, both_y.max()*1.10)

# Legend
plt.legend(loc='upper left')

for xi, label in events:
    yi = spx.loc[xi].values
    plt.scatter(xi, yi, 100, color = r_hex)
    plt.annotate(label, # Label
                 (xi,yi), # Coordinates 
                 xytext=(xi,yi+400), # Distance from text to points (x,y)
                 arrowprops=dict(facecolor='black', headwidth=4, width=1, headlength=6, shrink=0.2),
                 ha='left',va='top') # Horizontal/vertical alignment can be left, right or center
    
# Set up another axis
## Add second legend
ax2 = ax.twinx()
ax2.plot(x, y2, color=r_hex, linewidth  =3, linestyle = 'solid', label = 'Nikkei 225', alpha = 0.4)
ax2.legend(loc='lower right')

# Set up tick labels
## Indexing the X column values and converting to string format
event_tick_label = [x[0].strftime('%y-%m-%d') for x in events]
event_tick = pd.Index([x[0] for x in events])

## Create a list using list comprehension
## strftime convert Datetime format to String for the X tick labels
## We can rotate the labels to make it easier to read

# Combine both index together
xticks = x[range(0, len(x), 20)].union(event_tick)

ax.set_xticks(xticks)
ax.set_xticklabels([i.strftime('%y-%m-%d') for i in xticks], rotation=90)

for idx in ax.get_xticklabels():
    # idx is a matplotlib.text type
    if idx.get_text() in event_tick_label:
        idx.set_color(r_hex)
        idx.set_fontweight('bold')

plt.show()
```
